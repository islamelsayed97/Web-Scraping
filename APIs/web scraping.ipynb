### import needed libraries
**pandas** for saving the result in csv file.

**requests** for geting the content of the web page.

**BeautifulSoup** for organizing the data and extract what I want.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.programmableweb.com/category/all/apis'
```

pass the content to the BeautifulSoup that will make me can work with it.

```python
# empty list to add the data of each API to if
APIs = []
# loop to access each page
while True:
    response = requests.get(url)
    soup = BeautifulSoup(response.text)
    # get the table
    table = soup.find('table')
    # trs here is a list that contains the whole rows in the table
    trs = table.find_all('tr')
    # loop to access each row
    for tr in trs[1:-1]:
        # tds here is a list that contains the whole the whole cells in the row
        tds = tr.find_all('td')

        # first cell is the API name
        Name = tds[0].text
        # first cell contain on '<a> tag' where the link is exist in its 'href'
        Url = tds[0].find('a').get('href')
        # second cell is the API discription
        Discription = tds[1].text
        # third cell is the API category
        Category = tds[2].text
        # add the four variables to the APIs list as a dictionary
        APIs.append({'API Name':Name,
                     'API URL':Url,
                     'API Discription':Discription,
                     'API Category':Category})
    # get the link of the next page
    next_page = soup.find('a',{'title':'Go to next page'})
    
    # if there is an error in this part, this mean that the current page doesn't contain on a link to the next page.
    # there isn't next page when I'm in the last page, so the line code inside try will generate an error in the last page
    # so that I put it in the try statment, when the error happen I want to exit the loop because no pages more.
    try:
        # add the next page link to the essential link 
        url = 'https://www.programmableweb.com' + next_page.get('href')
        print(url)
    except Exception:
        break
```

```python
# make a data frame from the data
APIs_df = pd.DataFrame(APIs)
APIs_df
```

```python
# save the data as csv file
APIs_df.to_csv('APIs_df.csv', index = False)
```
